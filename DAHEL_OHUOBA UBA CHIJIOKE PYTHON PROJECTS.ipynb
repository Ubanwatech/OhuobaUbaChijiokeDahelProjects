```python
#FIBONACCI SEQUENCE
def generate_fibonacci(n):
    """Generate Fibonacci sequence up to n terms."""
    fib_sequence = []
    a, b = 0, 1
    for _ in range(n):
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

def main():
    # Prompt the user for the number of terms
    try:
        n = int(input("Enter the number of terms for the Fibonacci sequence: "))
        if n < 0:
            print("Please enter a non-negative integer.")
        else:
            fibonacci_sequence = generate_fibonacci(n)
            print(f"The first {n} terms of the Fibonacci sequence are: {fibonacci_sequence}")
    except ValueError:
        print("Invalid input. Please enter a valid integer.")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
#PALIDROME CHECKER
def is_palindrome(s):
    s = ''.join(okwu for okwu in s if okwu.isalnum()).lower()
    return s == s[::-1]

def main():
    words = input("Enter your words: ").split()
    
    for word in words:
        if is_palindrome(word):
            print(f"'{word}' is a palindrome.")
        else:
            print(f"'{word}' is not a palindrome.")

if __name__ == "__main__":
    main()
```

```python
#PRIME NUMBER CHECKER
def is_prime(num):
    """Check if a number is prime."""
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def main():
    # Get user input
    numbers = input('Enter your numbers separated by spaces: ')
    numbers = map(int, numbers.split())  # Convert input to a list of integers

    for number in numbers:
        if is_prime(number):
            print(f'{number} is a prime number')
        else:
            print(f'{number} is not a prime number')

if __name__ == "__main__":
    main()
```

```python
#BASIC FILE HANDLING
def create_text_file(filename, content):
    """Create a text file and write content to it."""
    with open(filename, 'w') as file:
        file.write(content)
    print(f"File '{filename}' created and content written.")

def count_file_contents(filename):
    """Count the number of words, lines, and characters in a text file."""
    with open(filename, 'r') as file:
        content = file.read()
        
        # Count lines
        lines = content.splitlines()
        line_count = len(lines)
        
        # Count words
        words = content.split()
        word_count = len(words)
        
        # Count characters
        char_count = len(content)
        
    return line_count, word_count, char_count

def main():
    # Define the filename and content
    filename = 'sample.txt'
    content = """Hello, world!
This is a sample text file.
It contains multiple lines and words.
Let's count them!"""

    # Create the text file and write content to it
    create_text_file(filename, content)

    # Count the contents of the file
    line_count, word_count, char_count = count_file_contents(filename)

    # Print the results
    print(f"File '{filename}' contains:")
    print(f"Lines: {line_count}")
    print(f"Words: {word_count}")
    print(f"Characters: {char_count}")

if __name__ == "__main__":
    main()
```

```python
#SIMPLE CALCULATOR. BETTER RUN WITH VS CODE
from tkinter import *

# Create the main window
root = Tk()
root.title("Simple Calculator")
root.geometry("400x600")
root.resizable(0, 0)

# Global variable to store the expression
expression = ""

# Function to update the expression in the input field
def btn_click(item):
    global expression
    expression += str(item)
    input_text.set(expression)

# Function to clear the input field
def bt_clear():
    global expression
    expression = ""
    input_text.set("")

# Function to evaluate the expression and display the result
def bt_equal():
    global expression
    try:
        result = str(eval(expression))
        input_text.set(result)
        expression = ""
    except Exception as e:
        input_text.set("Error")
        expression = ""

# Input field to display the expression/result
input_text = StringVar()
input_frame = Frame(root)
input_frame.pack(pady=10)

input_field = Entry(input_frame, textvariable=input_text, font=('Arial', 24), bd=10, insertwidth=4, width=14, borderwidth=4)
input_field.grid(row=0, column=0)

# Buttons for calculator
button_frame = Frame(root)
button_frame.pack()

buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    'C', '0', '=', '+'
]

row_val = 0
col_val = 0

for button in buttons:
    if button == '=':
        Button(button_frame, text=button, padx=20, pady=20, font=('Arial', 18), command=bt_equal).grid(row=row_val, column=col_val)
    elif button == 'C':
        Button(button_frame, text=button, padx=20, pady=20, font=('Arial', 18), command=bt_clear).grid(row=row_val, column=col_val)
    else:
        Button(button_frame, text=button, padx=20, pady=20, font=('Arial', 18), command=lambda item=button: btn_click(item)).grid(row=row_val, column=col_val)

    col_val += 1
    if col_val > 3:
        col_val = 0
        row_val += 1

# Start the GUI event loop
root.mainloop()
```

```python
#BASIC TIC TAC GAME. BEST RUN WITH VS CODE
import tkinter as tk
from tkinter import messagebox

class TicTacToe:
    def _init_(self, master): 
#What Does def _init_(self, master): Mean?
#What is def?
#In Python, def is a keyword that means you're defining a function or a method. 
# A method is just a function that belongs to a class.
#What is _init_?
#The _init_ method is special because it runs automatically when you create an object (or instance) of a class.
#  Think of it like the setup for your game or program. It prepares everything you need right when you start.
        self.master = master
        self.master.title("Tic Tac Toe")
        
        # Create a frame for the game
        self.frame = tk.Frame(self.master)
        self.frame.pack()

        # Initialize the game board
        self.board = [None] * 9
        self.current_player = "X"

        # Create buttons for the Tic Tac Toe grid
        self.buttons = []
        for i in range(3):
            row = []
            for j in range(3):
                button = tk.Button(self.frame, text="", font=('Arial', 24), width=5, height=2,
                                   command=lambda index=i*3+j: self.make_move(index))
                button.grid(row=i, column=j)
                row.append(button)
            self.buttons.append(row)

    def make_move(self, index):
        if not self.board[index] and not self.check_winner():
            self.board[index] = self.current_player
            self.buttons[index // 3][index % 3].config(text=self.current_player)
            if self.check_winner():
                messagebox.showinfo("Game Over", f"Player {self.current_player} wins!")
                self.reset_game()
            elif all(self.board):
                messagebox.showinfo("Game Over", "It's a draw!")
                self.reset_game()
            else:
                self.current_player = "O" if self.current_player == "X" else "X"

    def check_winner(self):
        winning_combinations = [
            (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Horizontal
            (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Vertical
            (0, 4, 8), (2, 4, 6)               # Diagonal
        ]
        for combo in winning_combinations:
            if self.board[combo[0]] == self.board[combo[1]] == self.board[combo[2]] != None:
                return True
        return False

    def reset_game(self):
        self.board = [None] * 9
        for button in self.buttons:
            for b in button:
                b.config(text="")
        self.current_player = "X"

if _name_ == "_main_":
    root = tk.Tk()
    game = TicTacToe(root)
   root.mainloop()
```

```python
#REMOVING HTML TAGS 
from bs4 import BeautifulSoup

def remove_html_tags(html_content):
    """
    Remove HTML tags from a given string.

    Parameters:
    html_content (str): The string containing HTML content.

    Returns:
    str: The cleaned string without HTML tags.
    """
    soup = BeautifulSoup(html_content, 'html.parser')
    return soup.get_text(strip=True)

# Example usage
if __name__ == "__main__":
    # Sample HTML content
    sample_html = "<div><p>This is a <strong>sample</strong> paragraph.</p></div>"
    
    # Remove HTML tags
    cleaned_text = remove_html_tags(sample_html)
    print(cleaned_text)  # Output: This is a sample paragraph.
```
